# DAY 5 — **Deep Learning &** **Backpropagation**

## 1. What is Deep Learning?
Deep Learning is a subset of machine learning that uses neural networks with multiple layers to learn complex patterns.
*   When neural networks have many hidden layers, they become deep.
*   These networks automatically learn features, instead of manually extracting them.

**Examples of deep learning applications:**


*   Self-driving cars
*   Face recognition
*   ChatGPT & LLMs
*   Medical image classification
*   Speech recognition

# 2. Architecture of a Deep Neural Network
**A basic deep neural network has:**
*   **Input Layer –** receives raw data
*   **Hidden Layers (multiple) –** learns complex patterns
*   **Output Layer –** final prediction (class, number, probability)

**Each neuron has:**
*   Weights (W)
*   Bias (b)
*   Activation function (ReLU, Sigmoid, etc.)

# 3. Forward Propagation (How prediction happens)
**Forward pass steps:**
*   Multiply inputs with weights
*   Add bias
*   Apply activation function
*   Pass values to next layer

**Example:**

z = W*x + b

a = activation(z)

This continues until output is produced.



# 4. Loss Function
Loss tells how far prediction is from the target.

**Common loss functions:**
*   MSE (Mean Squared Error) → regression
*   Binary Cross Entropy → binary classification
*   Categorical Cross Entropy → multi-class classification
*   Goal = minimize loss

# 5. Backpropagation
Backpropagation (backward propagation of errors) is the fundamental algorithm used to train neural networks in deep learning. It works by calculating the gradient of the loss function with respect to each weight and bias in the network, allowing an optimizer (like gradient descent) to update these parameters and minimize error.

Backpropagation = Backward pass of the network & Adjusting weights and biases to reduce loss.

**Process:**
1.   Compute loss
2.   Move backwards layer-by-layer
3.   Calculate gradients (∂Loss/∂Weights)
4.   Update weights using Gradient Descent: **W_new = W_old – learning_rate * gradient**

Repeat this for many epochs → the model learns better.

# 6. Optimization Algorithms
**Popular optimizers:**
*   SGD
*   Adam (most widely used)
*   RMSProp
*   Momentum

**Adam is used because:**

✔ Converges faster

✔ Works well for most deep learning tasks

# 7. Activation Functions
**ReLU -**  f(x) = max(0, x) Used in hidden layers.

**Sigmoid -** Converts values into probability between 0–1.

**Softmax -** Used in final layer for multi-class classification.

# 8. Hands-On Python Code (Deep Neural Network + Backprop)
Using TensorFlow/Keras.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Building a simple deep neural network
model = Sequential([
    Dense(64, activation='relu', input_shape=(10,)),   # Input + Hidden Layer 1
    Dense(32, activation='relu'),                      # Hidden Layer 2
    Dense(1, activation='sigmoid')                     # Output Layer
])

# Compile the model (define optimizer + loss)
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Dummy data (for explanation)
import numpy as np
X = np.random.rand(100, 10)  # 100 samples, 10 features
y = np.random.randint(0, 2, 100)

# Train the model
history = model.fit(X, y, epochs=10, batch_size=8)

# Evaluate
model.evaluate(X, y)
```
